```python
import numpy as np
import gradio as gr
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Collect FAQs (Pre-defined Knowledge Base)
faq_data = [
    {
        "question": "What is CodeAlpha?",
        "answer": "CodeAlpha is a software development company dedicated to driving innovation and providing hands-on internship programs across emerging technologies."
    },
    {
        "question": "How many tasks do I need to complete for the internship?",
        "answer": "To be eligible for the internship certificate, participants must complete a minimum of two or three tasks. Submitting only one task is considered incomplete."
    },
    {
        "question": "Where should I submit my completed tasks?",
        "answer": "A submission form link will be shared directly in your respective WhatsApp group. Make sure to follow the instructions carefully."
    },
    {
        "question": "What is the naming convention for the GitHub repository?",
        "answer": "You must upload your source code to a public GitHub repository named following this exact structure: CodeAlpha_ProjectName."
    },
    {
        "question": "What perks do interns receive upon successful completion?",
        "answer": "Interns receive an Internship Offer Letter, a QR-verified Completion Certificate, a Unique ID Certificate, a performance-based Letter of Recommendation, Job Opportunities, and Resume Building Support."
    }
]

# Extract questions and answers for processing
questions = [item["question"] for item in faq_data]
answers = [item["answer"] for item in faq_data]

# 2. Preprocess text and fit the TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(questions)

# 3. Match user questions and return responses
def chatbot_response(user_message, history):
    if not user_message.strip():
        return "Please ask a question regarding the CodeAlpha internship."

    # Transform user query using the same vectorizer
    query_vector = vectorizer.transform([user_message])

    # Calculate Cosine Similarity against all FAQ questions
    similarities = cosine_similarity(query_vector, tfidf_matrix).flatten()
    best_match_idx = np.argmax(similarities)
    highest_score = similarities[best_match_idx]

    # Check similarity confidence threshold (e.g., 0.2)
    if highest_score > 0.2:
        return answers[best_match_idx]
    else:
        return "I'm sorry, I couldn't find a direct answer to that. Please ask about the CodeAlpha internship overview, task submissions, repository naming, or certification perks!"

# 4. Create the Chat UI Interface using Gradio
demo = gr.ChatInterface(
    fn=chatbot_response,
    title="🤖 CodeAlpha FAQ Chatbot",
    description="Ask me anything about the CodeAlpha Artificial Intelligence Internship instructions, rules, perks, and submission guidelines!",
    examples=["What is CodeAlpha?", "How many tasks do I need to complete?", "Where do I submit my tasks?", "What are the internship perks?"],
    theme="soft"
)

# Launch the chatbot interface inline inside Google Colab
demo.launch(inline=True)
```
